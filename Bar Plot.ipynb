```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
data_filename = "amazon_data_18-11-19clean2.csv"
df = pd.read_csv(data_filename,delimiter=";")
df
```

```python
colors = ("red","blue")
plt.bar(df['name'],df['seller_rating'],alpha=0.5)
plt.title('BAR PLOT')
plt.xlabel('name')
plt.xticks(rotation=90)
plt.ylabel('stars')
plt.ylim(4.4,5.2)
plt.show()
```
